```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Max depth = 15

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20, 10))
df = pd.read_csv("ands.txt", sep=" ")
ax1.hist(df["depth"])
ax1.set_title("AND nodes, max_depth = 2")
print("Intersection", sum(df["intersection"]), len(df))

df = pd.read_csv("ors.txt", sep=" ")
ax2.hist(df["depth"], bins = 15)
ax2.set_title("OR nodes, max_depth = 2")
print("Intersection", sum(df["intersection"]), len(df))
```
